# 4. Hashing Vectorizer

Uses the hashing trick to map words to indices in a fixed-size vector.

## Advantages
- Low memory (no vocabulary storage).
- Scalable (online learning).

## Issues
- **Hash Collisions:** Multiple words mapping to the same index.
- **Non-reversible:** Can't go back from index to word.

```python
from sklearn.feature_extraction.text import HashingVectorizer
import pandas as pd

df_train = pd.read_csv('../Dataset/train.txt', sep=';', names=['text', 'emotion'])

# Fixed size vector (n_features is power of 2 usually)
vectorizer = HashingVectorizer(n_features=2**10) # 1024 features
X = vectorizer.transform(df_train['text'].iloc[:5])

print("Shape:", X.shape)
print("First document representation:\n", X.toarray()[0])
```
