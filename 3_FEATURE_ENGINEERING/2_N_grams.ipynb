# 2. N-grams

N-grams capture local context by considering sequences of N words.

## Types
- **Unigram (N=1):** Single words (standard BoW).
- **Bigram (N=2):** Pairs of consecutive words.
- **Trigram (N=3):** Triplets of consecutive words.

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

df_train = pd.read_csv('../Dataset/train.txt', sep=';', names=['text', 'emotion'])
data = df_train['text'].iloc[:5]

# Bigrams
vectorizer_bigram = CountVectorizer(ngram_range=(2, 2))
X_bigram = vectorizer_bigram.fit_transform(data)

print("Bigram Vocabulary (Top 10):", vectorizer_bigram.get_feature_names_out()[:10])

# Mixed (Unigrams + Bigrams)
vectorizer_mix = CountVectorizer(ngram_range=(1, 2))
X_mix = vectorizer_mix.fit_transform(data)

print("\nMixed Vocabulary Size:", len(vectorizer_mix.get_feature_names_out()))
```

## Use Cases
- Capturing phrases like "New York" or "not good" (negation handling).
