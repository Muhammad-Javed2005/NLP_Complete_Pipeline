# 3. TF-IDF

**Term Frequency - Inverse Document Frequency**

## Formula
- **TF:** Count of word in doc / Total words in doc.
- **IDF:** log(Total docs / Docs containing word).
- **TF-IDF:** TF * IDF.

## TF-IDF vs BoW
TF-IDF penalizes common words (like 'the') and highlights rare, distinctive words.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

df_train = pd.read_csv('../Dataset/train.txt', sep=';', names=['text', 'emotion'])

# TfidfVectorizer
tfidf = TfidfVectorizer()
X_tfidf = tfidf.fit_transform(df_train['text'].iloc[:10]) # First 10 docs

print("Shape:", X_tfidf.shape)
print("Sample Weights:\n", X_tfidf.toarray()[0])
```
