# 4. Stemming

Stemming reduces words to their root form (e.g., 'running' -> 'run'). Often heuristic-based.

## 1. Porter Stemmer
Older, faster, less aggressive.

## 2. Snowball Stemmer
Improved version of Porter (also called Porter2), supports multiple languages.

```python
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.tokenize import word_tokenize
import pandas as pd

porter = PorterStemmer()
snowball = SnowballStemmer('english')

words = ['running', 'runs', 'ran', 'easily', 'fairly']
```

```python
print("{:<15} {:<15} {:<15}".format("Word", "Porter", "Snowball"))
for w in words:
    print("{:<15} {:<15} {:<15}".format(w, porter.stem(w), snowball.stem(w)))
```

## Advantages & Drawbacks
**Advantages:**
- Reduces vocabulary size significantly.
- Fast to compute.

**Drawbacks:**
- Can produce non-words (e.g., 'easily' -> 'easili').
- Loss of meaning.
