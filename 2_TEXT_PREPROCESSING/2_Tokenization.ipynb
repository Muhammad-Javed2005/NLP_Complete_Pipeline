# 2. Tokenization

Tokenization is the process of breaking text into smaller units (tokens). We will cover:
1. Word Tokenization
2. Sentence Tokenization
3. N-gram Tokenization

```python
import pandas as pd
import nltk
# Ensure necessary NLTK data is downloaded
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')
    
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import ngrams
```

## Load Dataset

```python
df_train = pd.read_csv('../Dataset/train.txt', sep=';', names=['text', 'emotion'])
sample_text = df_train['text'].iloc[0]
print("Sample Text:", sample_text)
```

## 1. Word Tokenization

```python
def tokenize_words(text):
    return word_tokenize(text)

print("Word Tokens:", tokenize_words(sample_text))

df_train['word_tokens'] = df_train['text'].apply(tokenize_words)
```

## 2. Sentence Tokenization

```python
long_text = "This is sentence one. And this is sentence two! Finally, sentence three."
print("Long Text:", long_text)
print("Sentence Tokens:", sent_tokenize(long_text))
```

## 3. N-gram Tokenization

```python
def get_ngrams(text, n):
    tokens = word_tokenize(text)
    return list(ngrams(tokens, n))

print("Bigrams (N=2):", get_ngrams(sample_text, 2))
print("Trigrams (N=3):", get_ngrams(sample_text, 3))
```
